```python
#Ans1.
Matplotlib is a popular data visualization library for Python. It provides a wide variety of plotting options and customization features, making it a go-to tool for generating publication-quality graphs, charts, and figures.
Matplotlib is used for a range of applications, including data exploration, data analysis, and presentation of scientific findings.

The five plots are-
1.Line Plot: Used to represent the data points on a straight line. It is commonly used to show the trend of the data over time.
2.Scatter Plot: Used to represent the relationship between two variables. It is commonly used to identify the correlation between two variables.
3.Bar Plot: Used to represent the frequency of data values in a given set. It is commonly used to compare different categories or groups.
4.Histogram: Used to represent the frequency distribution of a variable. It is commonly used to visualize the distribution of data.
5.Pie Chart: Used to represent the proportion of different categories in a dataset. It is commonly used to show the relative sizes of different categories or groups.
```

```python
#Ans2.
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randn(100)
y = 2*x + np.random.randn(100)

plt.scatter(x, y)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of plots
fig, axs = plt.subplots(nrows=2, ncols=2)

# Plot 1: Top-left
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Plot 2: Top-right
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Plot 3: Bottom-left
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plot 4: Bottom-right
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Add a common x-label and y-label
fig.text(0.5, 0.04, 'X-axis', ha='center')
fig.text(0.04, 0.5, 'Y-axis', va='center', rotation='vertical')

# Add a main title
fig.suptitle('Four Line Plots')

# Show the plots
plt.show()
```

```python
#Ans5.
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

fig, ax = plt.subplots()

ax.boxplot([box1, box2])

ax.set_xticklabels(['Box 1', 'Box 2'])

ax.set_title('Box Plot')
ax.set_ylabel('Data')

plt.show()
```

```python
#Ans4.
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])


plt.bar(company,profit)
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title("Company Profit")
plt.show()
```
